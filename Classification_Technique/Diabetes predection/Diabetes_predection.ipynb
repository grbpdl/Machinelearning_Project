Diabetes Predictor using Vector Support Machine

Support vector machine is advanced form of logistics regression which have less anomalies and less margin and good accuracy

Support vector machine is a clasification technique

```python
#required dependencies
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#data collection and preprocessing and analysis
#loading datasets
data=pd.read_csv('/content/diabetes.csv')
```

```python
#seperating data and labels
x=data.drop(columns='Outcome',axis=1)
y=data['Outcome']
```

Data standardization
each features of data like gulcose ,insulin,age have their own ranges so with data standardization machine learning makes better predection


```python
scaler=StandardScaler()
scaler.fit(x)
standarized_data=scaler.transform(x)
```

```python
x=standarized_data
y=data['Outcome']
```

Train,test and split data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Traning the model

```python
classifier=svm.SVC(kernel='linear')
```

```python
#traning support vector machine classifier
classifier.fit(x_train,y_train)
```

Model Evaluation

Accuracy of model

```python
#accuracy on traning data
x_train_predection=classifier.predict(x_train)
tranning_data_accuracy=accuracy_score(x_train_predection,y_train)
print(tranning_data_accuracy)
```

```python
#accuracy on testing data
x_test_predection=classifier.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_predection,y_test)
print(testing_data_accuracy)
```

Makng a predecton system

```python
input_data=(4,110,92,0,0,37.6,0.191,30)
#changing input into numpy array
input_as_numpy_array=np.asarray(input_data)
#reshaping array
reshaped_data=input_as_numpy_array.reshape(1,-1)
#standardizing the data
standard_data=scaler.transform(reshaped_data)
print(standard_data)
predection=classifier.predict(standard_data)
if(predection==0):
  print("non diabetic")
else:
  print("diabetic")
```
